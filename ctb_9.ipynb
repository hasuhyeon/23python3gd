5. 3-6-9계임(3의 배수 판단): 3의 배수를
3으로 나누면 나머지가0이 되므로(1 mod 3)의_
값이 0아면 박수를 출력, 아니면 숫자를 출력하는 코드

```python
for num in range(1, 20):
 if num % 3 == 0:
    print("박수", end=' ')
 else:
    print(num, end=" ")
```

```python
num = 1
while num <= 20:
   if num % 3== 0:
     print(num, end=' ')
   else:
    print(num, end=' ')
    num += 1
```
